```python
Letters=["a","b","c","d","e"]
Letters
```

```python
#append
Letters=["a","b","c","d","e"]
Letters.append("f")
print(Letters)
```

```python
#extend
Letters.extend(["g","h"])
print(Letters)
```

```python
#insert
Letters.insert(5,"x")
print(Letters)
```

```python
#remove
Letters.remove("x")
print(Letters)
```

```python
#pop
pop_Element=Letters.pop(5)
print(pop_Element)
print(Letters)
```

```python
#Index
index=Letters.index("e")
print(index)
```

```python
#count
Count=Letters.count('h')
print(Count)
```

```python
#sort
Letters.sort()
Letters
```

```python
#reverse
Letters.reverse()
Letters
```
